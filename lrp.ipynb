```python
import matplotlib.pyplot as plt

import json

import data
```

```python
MODELS_FOLDER = "E:/Master/Avhandling/Models/models"

IMPL = "fixslice"
TARGET = "key"
BYTE = 1
ARCH = "zhang"
TRACE_START = 0
TRACE_END = 1000

MODEL_NAME = f"{IMPL}-{TARGET}-byte{BYTE}-{ARCH}-{TRACE_START}_{TRACE_END}"
```

```python
with open(f"{MODELS_FOLDER}/lrp/{MODEL_NAME}.json", 'r') as f:
    attr_info = json.load(f)



trace = data.trace_sample()

fig, axs = plt.subplots(nrows=2, ncols=1)
fig.set_figheight(8.0)
fig.set_figwidth(12.0)

axs[0].plot(attr_info["avg_attr"], 'magenta')
#axs[1].plot(attr_info["avg_abs_attr"], 'purple')
axs[1].plot(trace[0:1000])

axs[0].set_title("LRP attribution")
#axs[1].set_title("Abs LRP")
axs[1].set_title("Example Trace (raw)")

plt.show()
```
