```python
import matplotlib.pyplot as plt

import json

import data
```

```python
MODELS_FOLDER = "E:/Master/Avhandling/Models/models"

IMPL = "fixslice"
TARGET = "sbox"
BYTE = 1
ARCH = "zhang"
TRACE_START = 0
TRACE_END = 1000

MODEL1_NAME = f"{IMPL}-{TARGET}-byte{BYTE}-{ARCH}-{TRACE_START}_{TRACE_END}"
```

```python
IMPL = "fixslice"
TARGET = "sbox2"
BYTE = 1
ARCH = "zhang"
TRACE_START = 0
TRACE_END = 1000

MODEL2_NAME = f"{IMPL}-{TARGET}-byte{BYTE}-{ARCH}-{TRACE_START}_{TRACE_END}"
```

```python
with open(f"{MODELS_FOLDER}/lrp/{MODEL1_NAME}.json", 'r') as f:
    attr_info1 = json.load(f)

with open(f"{MODELS_FOLDER}/lrp/{MODEL2_NAME}.json", 'r') as f:
    attr_info2 = json.load(f)

trace = data.trace_sample()

fig, axs = plt.subplots(nrows=3, ncols=1)
fig.set_figheight(8.0)
fig.set_figwidth(12.0)

interval_start = 370
interval_end = 900

import numpy as np

axs[0].plot(np.arange(interval_start, interval_end), attr_info1["avg_attr"][interval_start:interval_end], 'blue', label='plaintext 0')
axs[0].plot(np.arange(interval_start, interval_end), attr_info2["avg_attr"][interval_start:interval_end], 'red', label='plaintext 1')
axs[1].plot(np.arange(interval_start, interval_end), attr_info1["avg_abs_attr"][interval_start:interval_end], 'purple', label='plaintext 0')
axs[1].plot(np.arange(interval_start, interval_end), attr_info2["avg_abs_attr"][interval_start:interval_end], 'magenta', label='plaintext 1')

axs[2].plot(np.arange(interval_start, interval_end), trace[interval_start:interval_end])

axs[0].legend()
axs[1].legend()

axs[0].set_title("LRP")
axs[1].set_title("Abs LRP")
axs[2].set_title("Example Trace")

plt.show()
```
