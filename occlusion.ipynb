```python
train_traces = train_traces[:, 0:1000]

train_std = train_traces.std()
train_mean = train_traces.mean()

norm_traces = (train_traces - train_mean) / train_std
```

```python
MODELS_FOLDER = "E:/Master/Avhandling/Models/models"

IMPL = "fixslice"
PRED_TARGET = "sbox"
TARGET_BYTE = 1
ARCH = "zhang"
INTERVAL_START = 0
INTERVAL_END = 1000

EPOCH = 54

MODEL_PATH = f"{MODELS_FOLDER}/{IMPL}-{PRED_TARGET}-byte{TARGET_BYTE}-{ARCH}-{INTERVAL_START}_{INTERVAL_END}/epoch{EPOCH}.pt"
```

```python
def occlusion_relevance(
        model : torch.nn.Module,
        input_trace : torch.Tensor,
        true_label : int,
        interval_start : int,
        interval_end : int,
):
    normal_score = model(input_trace)[0, true_label]

    occlusion_diffs = torch.zeros(input_trace.shape[-1])

    for pos in torch.arange(interval_start, interval_end):
        occluded = input_trace.clone()
        occluded[:, pos] = 0.

        occluded_score = model(occluded)[0, true_label]

        occlusion_diffs[pos] = normal_score - occluded_score
        

    return occlusion_diffs
```

```python
from models import *

model = torch.load(MODEL_PATH).cuda()

sample = 4444

trace = norm_traces[sample].cuda().unsqueeze(0)

label = train_sbox_output[sample, 0, TARGET_BYTE].long().item()
```

```python
INTERVAL_START = 0
INTERVAL_END = 1000

occlusion_scores = occlusion_relevance(
    model,
    trace,
    label,
    INTERVAL_START,
    INTERVAL_END,
)
```
